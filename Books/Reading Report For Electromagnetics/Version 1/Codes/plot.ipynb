```python
import matplotlib.pyplot as plt
import numpy as np
from sympy import *

x = symbols('x')

f = (x+2)*(x+1)*(x)*(x-1)
plot(f,(x,-3,2),ylim=(-2,3))
```

```python

f = (x)*(x-1)
plot(f,(x,-3,2),ylim=(-2,3))
```

```python
w = symbols('\omega')
z = 1/(1+w**2)**(1/2)
z
```

```python
h = lambdify(w,z,"numpy")
x = np.linspace(-10,10,100)
i = lambda t: t/(2*t)
plt.plot(x,h(x),x,i(x))
plt.xlabel(r'$\omega R_FC_F$')
plt.ylabel(r'$Z/R_F$')
```
