# Gamble Money Away

## Symbols
1. $p$: Probability to win in a round of bet.
2. $p_i$: Probability to lose, owning $i$ money. 
3. $L$: Total amount of money.
4. $M$: Money of one bettor.

```python
from sympy import *
M, p, q, L = symbols('M, p, q, L')

q = 1-p
pi = ((q/p)**M-(q/p)**L)/(1-(q/p)**L)
pi
```

## $p_i-M$ Relationship at Given $L,\ p$
### $p>0.5$ Greater Opportunity to Win

```python
plot(pi.subs({p:0.51,L:100}),(M,0,100),ylabel = r'$p_i,\ p=0.51$',xlabel = 'M')
```

### $p<0.5$ Less Opportunity to Win

```python
plot(pi.subs({p:0.49,L:100}),(M,0,100),ylabel = r'$p_i,\ p=0.49$',xlabel = 'M')
```

## $p_{50}-L$ Relationship at Given $M,\ p$
### $p>0.5$ Greater Opportunity to Win

As show below, the probability to lose $(p_{50})$ has a limit much lower than 1 which is under $14\%$ in this example. This result implies that if one has 50 fixed amount of money, he can still gamble without much opportunity to lose as long as the $p$ is greater than one half.

```python
plot(pi.subs({p:0.51,M:50}),(L,50,500),ylabel = r'$p_i,\ p=0.51$',xlabel = 'L')
```

### $p<0.5$ Less Opportunity to Win
Now the limit of $p_{50}$ is $1$ which can also be reach quickly.

```python
plot(pi.subs({p:0.49,M:50}),(L,50,500),ylabel = r'$p_i,\ p=0.49$',xlabel = 'L')
```
