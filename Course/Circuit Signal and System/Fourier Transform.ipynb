```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
e,n = symbols('E,n')
f = e/(2*pi*n)
f
```

```python
x = [i for i in range(-10,10)]
x.remove(0)
x = np.array(x)
```

```python
g = lambdify(n,f.subs({e:1}),'numpy')
g(x)
```

```python
plt.vlines(x,0,g(x))
plt.grid(True)
```
