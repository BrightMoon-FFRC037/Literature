```python
from sympy import *

init_printing()
# Define the time-domain variable and Laplace variable
t, s = symbols('t s')

# Define the function in the time domain
f = exp(-t) * Heaviside(t)

# Compute the Laplace Transform
F = laplace_transform(f, t, s)
F
```

```python

# Define the Laplace domain variable and time-domain variable
s, t = symbols('s t')

# Define the function in the Laplace domain
F = 1 / (s + 1)

# Compute the inverse Laplace Transform
f = inverse_laplace_transform(F, s, t)
f
```

```python
A,M,c,k,p,w = symbols('A,M,c,k,p,\omega')
F = (A*w)/((p**2+w**2)*(M*p**2+c*p+k))
F
```
