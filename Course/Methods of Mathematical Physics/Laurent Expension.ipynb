```python
import numpy as np
import matplotlib.pyplot as plt

# Define a grid in the complex plane
x = np.linspace(-2, 2, 400)
y = np.linspace(-2, 2, 400)
X, Y = np.meshgrid(x, y)
Z = X + 1j * Y

# Function 1/(1-z)
f = 1 / (1 - Z)

# Calculate magnitude and phase
magnitude = np.abs(f)
phase = np.angle(f)

# Create a 3D plot for magnitude and phase
fig = plt.figure(figsize=(14, 6))

# 3D plot for magnitude
ax1 = fig.add_subplot(121, projection='3d')
ax1.plot_surface(X, Y, magnitude, cmap='inferno', edgecolor='none')
ax1.set_title(r'3D Magnitude of $\frac{1}{1 - z}$')
ax1.set_xlabel('Re(z)')
ax1.set_ylabel('Im(z)')
ax1.set_zlabel('Magnitude')

# 3D plot for phase
ax2 = fig.add_subplot(122, projection='3d')
ax2.plot_surface(X, Y, phase, cmap='twilight', edgecolor='none')
ax2.set_title(r'3D Phase of $\frac{1}{1 - z}$')
ax2.set_xlabel('Re(z)')
ax2.set_ylabel('Im(z)')
ax2.set_zlabel('Phase')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the grid for the complex plane
x = np.linspace(-2, 2, 400)
y = np.linspace(-2, 2, 400)
X, Y = np.meshgrid(x, y)
Z = X + 1j * Y  # Complex plane grid

# Compute the modulus of cosh(Z)
modulus_cosh_z = np.abs(np.cosh(Z))

# Plot the modulus
plt.figure(figsize=(8, 6))
plt.contourf(X, Y, modulus_cosh_z, levels=50, cmap='viridis')
plt.colorbar(label=r'$|\cosh(z)|$')
plt.xlabel('Re(z)')
plt.ylabel('Im(z)')
plt.title(r'Modulus of $\cosh(z)$ in the Complex Plane')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the grid in the complex plane
x = np.linspace(-1, 1, 400)  # Real part range
y = np.linspace(-4, 4, 400)  # Imaginary part range
X, Y = np.meshgrid(x, y)
Z = X + 1j * Y

# Compute the modulus of cosh(z)
cosh_z = np.cosh(Z)
modulus_cosh_z = np.abs(cosh_z)

# Plotting the 3D surface
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, modulus_cosh_z, cmap='viridis', edgecolor='none')

# Labeling the axes
ax.set_xlabel("Re(z)")
ax.set_ylabel("Im(z)")
ax.set_zlabel(r'$|\cosh(z)|$')
ax.set_title(r"3D Plot of $|\cosh(z)|$ in the Complex Plane")

# Show the plot
plt.show()
```
