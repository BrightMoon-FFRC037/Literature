```python
from sympy import *
```

```python
# Define the variable
x = symbols('x')

# Define the Gamma and Psi functions in SymPy
gamma_expr = gamma(x)
psi_expr = polygamma(0, x)  # polygamma(0, x) is the digamma (psi) function

# Plotting the Gamma and Psi functions
p1 = plot(gamma_expr, (x, 0.1, 5), show=False, line_color='blue', label=r'$\Gamma(x)$')
p2 = plot(psi_expr, (x, 0.1, 5), show=False, line_color='red', label=r'$\psi(x)$')

# Set up the plot for displaying both
p1.extend(p2)  # Combine plots
p1.title = "Gamma and Psi Functions"
p1.xlabel = "x"
p1.ylabel = "Function Value"
p1[0].line_color = 'blue'
p1[1].line_color = 'red'
p1.show()
```
