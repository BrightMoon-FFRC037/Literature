## An Example

```python
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np

# Configure SymPy to use pretty printing in the notebook
sp.init_printing()

# Define symbols for function input
x = sp.symbols('x')
```

```python
# Define a symbolic function
f = sp.sin(x) + x**2

# Display the function in its symbolic form
f
```

```python
# Convert the symbolic function to a numerical one using lambdify
f_num = sp.lambdify(x, f, 'numpy')

# Create an array of values for plotting
x_vals = np.linspace(-5, 5, 400)  # Define the range of x-values

# Evaluate the function at those points
y_vals = f_num(x_vals)

# Plot the function
plt.plot(x_vals, y_vals)
plt.title('Plot of the Function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.show()
```

```python
from sympy.parsing.sympy_parser import parse_expr

# Input function as a string
function_str = input("Enter a function of x (e.g., sin(x) + x**2): ")

# Parse the input string into a symbolic expression
f = parse_expr(function_str)

# Convert to a numerical function
f_num = sp.lambdify(x, f, 'numpy')

# Plot
y_vals = f_num(x_vals)
plt.plot(x_vals, y_vals)
plt.title(f'Plot of {function_str}')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.show()
```

## Walk Through SymPy

```python
from sympy import *
x, y, z = symbols('x y z')
init_printing()
```

```python
expr = cos(x) + 1
expr.subs(x,y)
```

```python
expr.subs(x,0)
```

```python
expr = x**y
expr
```

```python
expr = expr.subs(y, x**y)
expr
```

```python
expr = sin(2*x) + cos(2*x)
expand_trig(expr)
```

```python
expr.subs(sin(2*x), 2*sin(x)*cos(x))
```

```python
expr = x**3 + 4*x*y - z
expr.subs([(x,2),(y,4),(z,0)])
```

```python
expr = x**4 - 4*x**3 + 4*x**2 - 2*x + 3
replacements = [(x**i, y**i) for i in range(5) if i % 2 == 0]
replacements
expr.subs(replacements)
```

```python
str_expr = 'x**2 + 3*x - 1/2'
expr = sympify(str_expr)
expr
```
