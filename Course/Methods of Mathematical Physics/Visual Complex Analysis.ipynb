```python
from matplotlib import pyplot as plt
import numpy as np
from sympy import *

x, y, u, v, z, w = symbols('x, y, u, v, z, w')
init_printing()
```

$$
w = z^2
$$

```python
u = x**2 - y**2
v = 2*x*y
(u,v)
```

```python
u_p, v_p = [lambdify([x, y], u, 'numpy'), lambdify([x, y], v, 'numpy')] 
t = 8
X, Y = np.meshgrid(np.linspace(-t,t,2*t),np.linspace(-t,t,2*t)) 
U, V = u_p(X,Y), v_p(X,Y)
plt.quiver(X,Y,U,V, scale_units = 'xy', angles = 'xy', color = 'royalblue', width = 2e-3)
```

```python
X, Y = np.meshgrid(np.linspace(-t,t,30*t),np.linspace(-t,t,30*t)) 
U, V = u_p(X,Y), v_p(X,Y)
plt.contour(X,Y,Y,levels=10,colors = 'red') # x axis
plt.contour(X,Y,X,levels=10,colors = 'blue')# y axis
```

```python
plt.contour(X,Y,V,levels=10,colors = 'red') # transformed x axis
plt.contour(X,Y,U,levels=10,colors = 'blue')# transformed y axis
plt.contour(X,Y,Y,levels=10,colors = 'orange',alpha = 0.2) # transformed x axis
plt.contour(X,Y,X,levels=10,colors = 'green', alpha = 0.2)# transformed y axis
```

```python
t=2
R, Θ = np.meshgrid(np.linspace(0,t,200*t),np.linspace(0,np.pi/2,100))
X = R*np.cos(Θ)
Y = R*np.sin(Θ)

lev_r = np.linspace(0,t,4*t)
lev_θ = np.linspace(0,np.pi/2,7)
plt.contour(X,Y,R,levels = lev_r,colors = 'red')
plt.contour(X,Y,Θ,levels = lev_θ,colors='blue')
```

```python
r, θ = R**2, Θ*2
X = r*np.cos(θ)
Y = r*np.sin(θ)
lev_r = lev_r**2
lev_θ = lev_θ*2
plt.contour(X,Y,R,levels = lev_r,colors = 'red')
plt.contour(X,Y,Θ,levels = lev_θ,colors='blue')
```

```python
u = cosh(y)*cos(x)
v = -sinh(y)*sin(x)
(u,v)
```

```python
u_p, v_p = [lambdify([x, y], u, 'numpy'), lambdify([x, y], v, 'numpy')] 
t = 4
X, Y = np.meshgrid(np.linspace(-t,t,10*t),np.linspace(-2*t,2*t,10*t)) 
U, V = u_p(X,Y), v_p(X,Y)
plt.quiver(X,Y,U,V, scale_units = 'xy', angles = 'xy', color = 'royalblue', width = 2e-3)
```
