# Visualization
## Numbers
Numbers can be represented in the form of scatters and vectors.

```python
import matplotlib.pyplot as plt

# Define complex numbers
z1 = 2 + 3j
z2 = -1 - 4j
z3 = 1 + 1j

# Extract real and imaginary parts
reals = [z1.real, z2.real, z3.real]
imags = [z1.imag, z2.imag, z3.imag]
labels = ['z1', 'z2', 'z3']

# Create a scatter plot
plt.figure(figsize=(6, 6))
plt.scatter(reals, imags, color='blue')

# Annotate the points
for i, label in enumerate(labels):
    plt.text(reals[i], imags[i], label, fontsize=12, ha='right')

# Set labels and title
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.xlabel('Real Part')
plt.ylabel('Imaginary Part')
plt.title('Complex Numbers in the Complex Plane')
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

## Functions

```python
import matplotlib.pyplot as plt
import numpy as np

def f(z):
    return z**2

real_vals = np.linspace(-2,2,400)
imag_vals = np.linspace(-2,2,400)
real_grid, imag_grid = np.meshgrid(real_vals, imag_vals)
complex_grid = real_grid + 1j*imag_grid
result = f(complex_grid)

plt.imshow(np.abs(result), extent=(-2,2,-2,2), origin='lower')
plt.colorbar()
plt.title('Magnitude of f(z)')
plt.xlabel('Real Part')
plt.ylabel('Imaginary Part')
plt.show()
```
