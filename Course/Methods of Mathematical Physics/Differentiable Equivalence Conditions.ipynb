# Differentiable Equivalence Conditions
## Preliminary and Representations
### Preparation for Visualization

```python
import matplotlib.pyplot as plt
import numpy as np
from sympy import *
origin = np.zeros(2)
θ, k = symbols('θ, k')
du, uv, dx, dy = symbols('du, dv, dx, dy')
```


### In the Aspect of Complex Number
Let $z=x+iy$, $w=u+iv$, and $w=f(z)$
$$
w=f(z)
$$
The function can also be expressed in the form of
$$
u+iv=u(x+iy)+v(x+iy)
$$
$$
f'(z)=\lim_{\Delta z \to 0} \frac{\Delta w}{\Delta z}
$$
The derivative can also be expressed in the form of
$$
\lim_{\Delta z \to 0} \frac{\Delta u + i \Delta v}{\Delta x + i\Delta y}
$$

### In the Aspect of Binary Mapping
Since complex numbers is equivalent to two dimensional vectors, we can derive the counterpart of the complex function in the context of binary mapping.
$$
\begin{cases}
u=u(x,y)\\
v=v(x,y)
\end{cases}
$$
The Jacobian Matrix ($J$) is
$$
\left(
\begin{matrix}
    \Delta u\\
    \Delta v
\end{matrix}
\right)
=
\left(
\begin{matrix}
    u_x& u_y\\
    v_x& v_y
\end{matrix}
\right)
\left(
\begin{matrix}
    \Delta x\\
    \Delta y
\end{matrix}
\right)
$$
$$
J=
\left(
\begin{matrix}
    u_x& u_y\\
    v_x& v_y
\end{matrix}
\right)
$$
which means
$$
\begin{cases}
\Delta u=u_x\Delta x+u_y\Delta y\\
\Delta v=v_x\Delta x+v_y\Delta y
\end{cases}
$$


```python
du = k*(cos(θ)*dx-sin(θ)*dy)
dv = k*(sin(θ)*dx+cos(θ)*dy)

du = lambdify([dx,dy,k,θ],du,'numpy')
dv = lambdify([dx,dy,k,θ],dv,'numpy')
dx = lambdify(dx,dx,'numpy')
dy = lambdify(dy,dy,'numpy')
```

```python
du_1 = du(3,1,1.5,np.pi/3)
dv_1 = dv(3,1,1.5,np.pi/3)
dx_1 = dx(3)
dy_1 = dy(1)
```

```python
reals = [dx_1,du_1]
imags = [dy_1,dv_1]

plt.quiver(origin, origin, reals, imags, angles='xy', scale_units='xy', scale=1, color='royalblue')

t=8
plt.xlim(-t, t)
plt.ylim(-t, t)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.xlabel('Real Part')
plt.ylabel('Imaginary Part')
plt.title('Complex Numbers in the Complex Plane')

plt.show()
```


## From Differentiable to Cauchy-Riemann Equation
The existence of derivative has the following constrains:
> $Arg(\Delta z)-Arg(\Delta w)$ is a constant
> 
> $|\Delta z|=k|\Delta w|$ where $k$ is a constant

The corresponding constrains for the binary mapping are such that:
> $J$ is **a rotational transformation** 
> 
> followed by **a stretching transformation** 
> 
> ($J$ can be decomposed with them)

Hence:
$$
J=
k
\left(
\begin{matrix}
    \cos \theta& -\sin \theta\\
    \sin \theta& \cos \theta
\end{matrix}
\right)
=
\left(
\begin{matrix}
    k\cos \theta& -k\sin \theta\\
    k\sin \theta& k\cos \theta
\end{matrix}
\right)
=
\left(
\begin{matrix}
    u_x& u_y\\
    v_x& v_y
\end{matrix}
\right)
$$
Where $det(J)=k^2$.

And therefore the Cauchy-Riemann Condition is obvious:
$$
u_x=k\cos \theta=v_y\\
v_x=k\sin \theta=-u_y
$$
## And Vice Versa !
From Cauchy-Riemann Equation we get
$$
u_x=v_y\\
v_x=-u_y
$$
$$
\left(
\begin{matrix}
    u_x& u_y\\
    v_x& v_y
\end{matrix}
\right)=J=
\left(
\begin{matrix}
    a& -b\\
    b& a
\end{matrix}
\right)=
k
\left(
\begin{matrix}
    \cos \theta& -\sin \theta\\
    \sin \theta& \cos \theta
\end{matrix}
\right)
$$
where $k=\sqrt{a^2+b^2}$

This kind of matrix can be decompose in the following pattern:

>the transformation $J$ (Jacobian Matrix) is the composition of 
>
>a **rotational transformation** 
>
>and **a stretching transformation**

Thus following requirements will be met:
> $Arg(\Delta z)-Arg(\Delta w)$ is a constant
>
> $|\Delta z|=k|\Delta w|$ where $k$ is a constant

Hence the existence $
\lim\frac{\Delta w}{\Delta z}=\lim \frac{\Delta u + i \Delta v}{\Delta x + i\Delta y}$ and the derivative $f'(z)$.
