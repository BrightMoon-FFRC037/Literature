```python
from sympy import *
```

```python
a,b,k1,k2,c0,t = symbols('A,B,k_1,k_2,C_0,t')
para={
    k1:2,
    k2:1,
    c0:10
}
```

```python
ca = c0*exp(-k1*t)
ca
```

```python
cb = c0*(1-exp(-k1*t))
cb
```

## Concentration

```python
plot(ca.subs(para),cb.subs(para),(t,0,3))
```

## Speed

```python
va=-k1*c0*exp(-k1*t)
vb=-va
plot(va.subs(para),vb.subs(para),(t,0,3))
```

```python
cae = c0*((k1/(k1+k2))*exp(-(k1+k2)*t)+k2/(k1+k2))
cbe = c0*((-k1/(k1+k2))*exp(-(k1+k2)*t)+k1/(k1+k2))
plot(cae.subs(para),cbe.subs(para),(t,0,3))
```

```python
vae = diff(cae,t)
vbe = diff(cbe,t)
plot(vae.subs(para),vbe.subs(para),(t,0,3))
```
