# Poisson Distribution-Seeing Photons

```python
from sympy import *
import matplotlib.pyplot as plt
import numpy as np
k, α, λ, P = symbols('k, α, λ, P')
P = (α*λ)**k*exp(-α*λ)/factorial(k)
P
```

```python
tmp = lambdify([k, α, λ],P)
discrete_k = range(0,30,1)
discrete_P = [tmp(ki, 0.1, 15) for ki in discrete_k]
plt.bar(discrete_k, discrete_P)
```

```python
theta = 2
res = []
discrete_lambda = range(50)
for lamd in discrete_lambda:
    discrete_k = range(theta, 50, 1)
    discrete_P = [tmp(ki, 0.1, lamd) for ki in discrete_k]
    res.append(sum(discrete_P))
plt.plot(discrete_lambda, res)
plt.title(f'θ={theta}')
plt.xlabel('λ')
plt.ylabel(r'$P_{see}$')
```
