```python
from sympy import *
import numpy as np 
import matplotlib.pyplot as plt
h,oh,m,n,cl,ka,kw,c,v,v0=symbols('[H^{+}],[OH^{-}],[Tris],[TrisH^{+}],[Cl_{-}],Ka,Kw,c,V,V_0')
init_printing()
```

```python
eq1 = Eq(kw,h*oh)
eq1
```

```python
eq2=Eq(ka,h*m/n)
eq2
```

```python
#eq3=Eq(cl+oh,h+n)
eq3=Eq(cl,h+n)
eq3
```

```python
eq4=Eq(cl,c*v/(v+v0))
eq4
```

```python
eq5=Eq(m+n,c*v0/(v+v0))
eq5
```

```python
solution=solve((eq2,eq3,eq4,eq5),(h,m,n,cl))
solution
```
