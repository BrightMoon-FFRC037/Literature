# Central Limit Theorem with Random Walk
# Preliminaries

```python
from sympy import *
import numpy as np 
import matplotlib.pyplot as plt
n, p, L = symbols('n, p, L')
init_printing(use_latex=True)
means = []
variances = []
```

# An Unbiased 1-D Random Walker
$$
p_{left}=p_{right}=0.5
$$

```python
n = 1 # a walker will take n steps
res = []
for _ in range(5000): # 5000 random walkers
    position = 0 # start from 0
    for steps in range(n):
        dice = np.random.random_sample()  # a random number in [0,1)
        if dice>0.5:
            position += 1
        else:
            position -= 1
    res.append(position)
plt.hist(res, rwidth=0.9, bins=2)
plt.title(f'Unbiased Walker n={n}')
print(f'mean = {np.mean(res):.3f}L; variance = {np.var(res):.3f}L^2')
means.append(np.mean(res))
variances.append(np.var(res))
plt.show()
```

# Biased Walkers
$$
p_{forward}=0.7\\
p_{backward}=0.3
$$

```python
n = 1 # a walker will take n steps
res = []
for _ in range(5000): # 5000 random walkers
    position = 0 # start from 0
    for steps in range(n):
        dice = np.random.random_sample()  # a random number in [0,1)
        if dice>0.3: # 1-0.3= 70% forward
            position += 1
        else: # 0.3 = 30% backward
            position -= 1
    res.append(position)
plt.hist(res, rwidth=0.9, bins=2)
plt.title(f'Unbiased Walker n={n}')
print(f'mean = {np.mean(res):.3f}L; variance = {np.var(res):.3f}L^2')
means.append(np.mean(res))
variances.append(np.var(res))
plt.show()
```
