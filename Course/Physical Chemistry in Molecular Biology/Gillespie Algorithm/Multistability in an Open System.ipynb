```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def relaxation_time(k_total):
    # Set the rate parameter (Î»)
    lambda_param = k_total
    # Generate a random number
    random_number = np.random.exponential(scale=1/lambda_param)
    return random_number
k1,k2,k3=3,100,0.0046
```

# (b)

```python
state={'X':400,'Y':0}
time_trace=[0]
time_duration=0
y_num=[0]
for i in range(10000):
    k_to_y = state['X']*k1+state['X']*state['Y']*(state['Y']-1)*k3
    k_to_x = state['Y']*k2
    k_tot=k_to_x+k_to_y
    time_duration+=relaxation_time(k_tot)
    time_trace.append(time_duration)
    lucky_num = np.random.uniform(0,1)
    if lucky_num < k_to_x/k_tot:
        state['X']+=1
        state['Y']-=1
    else: 
        state['X']-=1
        state['Y']+=1
    y_num.append(state['Y'])
```

```python
plt.plot(time_trace,y_num)
```

The trend is the increasing of Y over time, despite the fluctuations and finally reaches a steady state.

```python
num_of_y_in_steady_state=[]
for j in range(10000):
    state={'X':400,'Y':0}
    time_trace=[0]
    time_duration=0
    y_num=[0]
    for i in range(10000):
        k_to_y = state['X']*k1+state['X']*state['Y']*(state['Y']-1)*k3
        k_to_x = state['Y']*k2
        k_tot=k_to_x+k_to_y
        time_duration+=relaxation_time(k_tot)
        time_trace.append(time_duration)
        lucky_num = np.random.uniform(0,1)
        if lucky_num < k_to_x/k_tot:
            state['X']+=1
            state['Y']-=1
        else: 
            state['X']-=1
            state['Y']+=1
        y_num.append(state['Y'])
    num_of_y_in_steady_state.append(np.mean(y_num[-10:]))
```

```python
plt.hist(num_of_y_in_steady_state,bins=40)
```
