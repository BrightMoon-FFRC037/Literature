```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## EGFP (Enhanced Green Fluorescent Protein) Excitation and Emission Spectra
The data source is [fpbase](https://www.fpbase.org/protein/egfp/).

```python
data = pd.read_csv('./fpbase_spectra.csv')
wavelength = np.array(data['wavelength'])
excitation = np.array(data['EGFP ex'])
emission = np.array(data['EGFP em'])
plt.plot(wavelength,excitation,label='Excitation')
plt.plot(wavelength,emission,label='Emission')
plt.legend()
plt.xlabel('Wavelength nm')
plt.ylabel('Normalized Brightness')
```

## With Optical Filters
The data source is [Chroma](https://www.chroma.com/products/sets/49002-et-egfp-fitc-cy2).

```python
data_et470 = pd.read_table('./ET470.txt')
data_et525 = pd.read_table('./ET525.txt')
data_t495 = pd.read_table('./T495.txt')
wave_et470 = np.array(data_et470['300.00'][:400])
wave_et525 = np.array(data_et525['300.50'][:800])
wave_t495 = np.array(data_t495['380.00'][:320])
y_et470 = np.array(data_et470['5.5893900707815E-7'][:400])
y_et525 = np.array(data_et525['9.5000001465451E-7'][:800])
y_t495 = np.array(data_t495['8.0339203123003E-5'][:320])
plt.fill_between(wavelength,excitation,label='Excitation',color='blue',linestyle='--',alpha=0.4)
plt.fill_between(wavelength,emission,label='Emission',color='green',linestyle='--',alpha=0.4)
plt.xlabel('Wavelength (nm)')
plt.ylabel('Normalized Brightness')
plt.legend()
ax2 = plt.gca().twinx()
ax2.plot(wave_et470,y_et470,label='Excitation Filter (ET470)',linewidth=2,color='blue')
ax2.plot(wave_et525,y_et525,label='Emission Filter (ET525)',linewidth=2,color='green')
ax2.plot(wave_t495,y_t495,label='Beam Splitter (T495)',linewidth=4,color='red',linestyle='--')
plt.ylabel('Percentage of Transmission (%T)')
plt.legend()
plt.xlim([400,600])
```

## Counter Part for ECFP (Enhanced Cyan Fluorescent Protein)
Data source at [Chroma](https://www.chroma.com/products/sets/49001-et-ecfp?fluorochromes=840#legend-selector1) too.

```python
data_ex = pd.read_table('./ECFP_EX.txt')
data_em = pd.read_table('./ECFP_EM.txt')
wave_ex = data_ex['377.00']
excitation = data_ex['0.3589999973774']
wave_em = data_em['447.00']
emission = data_em['0.02280000038445']

data_et436 = pd.read_table('./ET436.txt')
data_et480 = pd.read_table('./ET480.txt')
data_t455 = pd.read_table('./T455.txt')
wave_et436 = np.array(data_et436['200.00'][:])
wave_et480 = np.array(data_et480['300.50'][:])
wave_t455 = np.array(data_t455['380.00'][:])
y_et436 = np.array(data_et436['9.9999999392253E-9'][:])
y_et480 = np.array(data_et480['9.5000001465451E-7'][:])
y_t455 = np.array(data_t455['0.76629102230072'][:])
plt.fill_between(wave_ex,excitation,label='Excitation',color='blue',linestyle='--',alpha=0.4)
plt.fill_between(wave_em,emission,label='Emission',color='cyan',linestyle='--',alpha=0.4)
plt.xlabel('Wavelength (nm)')
plt.ylabel('Normalized Brightness')
plt.legend()
ax2 = plt.gca().twinx()
ax2.plot(wave_et436,y_et436,label='Excitation Filter (ET436)',linewidth=2,color='blue')
ax2.plot(wave_et480,y_et480,label='Emission Filter (ET480)',linewidth=2,color='cyan')
ax2.plot(wave_t455,y_t455,label='Beam Splitter (T455)',linewidth=2,color='red',linestyle='--')
plt.ylabel('Percentage of Transmission (%T)')
plt.legend()
plt.xlim([380,550])
```

## EGFP with ECFP ?

```python
data = pd.read_csv('./fpbase_spectra.csv')
wavelength = np.array(data['wavelength'])
excitation = np.array(data['EGFP ex'])
emission = np.array(data['EGFP em'])
plt.fill_between(wavelength,excitation,label='EGFP Excitation',color='blue',linestyle='--',alpha=0.4)
#plt.fill_between(wavelength,emission,label='EGFP Emission',color='green',linestyle='--',alpha=0.4)
plt.xlabel('Wavelength (nm)')
plt.ylabel('Normalized Brightness')
plt.legend()
ax2 = plt.gca().twinx()
ax2.plot(wave_et436,y_et436,label='ECFP Excitation Filter',linewidth=2,color='blue')
#ax2.plot(wave_et480,y_et480,label='ECFP Emission Filter',linewidth=2,color='cyan')
ax2.plot(wave_t455,y_t455,label='ECFP Beam Splitter',linewidth=2,color='red',linestyle='--')
plt.legend()
plt.xlim([400,600])
```

### EGFP Emission Does not Come into Eyepiece Obviously

```python
data = pd.read_csv('./fpbase_spectra.csv')
wavelength = np.array(data['wavelength'])
excitation = np.array(data['EGFP ex'])
emission = np.array(data['EGFP em'])
#plt.fill_between(wavelength,excitation,label='EGFP Excitation',color='blue',linestyle='--',alpha=0.4)
plt.fill_between(wavelength,emission,label='EGFP Emission',color='green',linestyle='--',alpha=0.4)
plt.xlabel('Wavelength (nm)')
plt.ylabel('Normalized Brightness')
plt.legend()
ax2 = plt.gca().twinx()
#ax2.plot(wave_et436,y_et436,label='ECFP Excitation Filter',linewidth=2,color='blue')
ax2.plot(wave_et480,y_et480,label='ECFP Emission Filter',linewidth=2,color='cyan')
ax2.plot(wave_t455,y_t455,label='ECFP Beam Splitter',linewidth=2,color='red',linestyle='--')
plt.legend()
plt.xlim([400,600])
```

### However EGFP Beam will Excite ECFP Considerably

```python
excitation = data_ex['0.3589999973774']
emission = data_em['0.02280000038445']

plt.fill_between(wave_ex,excitation,label='ECFP Excitation',color='blue',linestyle='--',alpha=0.4)
#plt.fill_between(wave_em,emission,label='ECFP Emission',color='cyan',linestyle='--',alpha=0.4)
plt.xlabel('Wavelength (nm)')
plt.ylabel('Normalized Brightness')
plt.legend()
ax2 = plt.gca().twinx()
ax2.plot(wave_et470,y_et470,label='EGFP Excitation Filter',linewidth=2,color='blue')
#ax2.plot(wave_et525,y_et525,label='EGFP Emission Filter',linewidth=2,color='green')
ax2.plot(wave_t495,y_t495,label='EGFP Beam Splitter',linewidth=4,color='red',linestyle='--')
plt.ylabel('Percentage of Transmission (%T)')
plt.legend()
plt.xlim([380,600])
```

### And ECFP Emission Passes Through EGFP Filter Easily

```python
excitation = data_ex['0.3589999973774']
emission = data_em['0.02280000038445']

#plt.fill_between(wave_ex,excitation,label='ECFP Excitation',color='blue',linestyle='--',alpha=0.4)
plt.fill_between(wave_em,emission,label='ECFP Emission',color='cyan',linestyle='--',alpha=0.4)
plt.xlabel('Wavelength (nm)')
plt.ylabel('Normalized Brightness')
plt.legend()
ax2 = plt.gca().twinx()
#ax2.plot(wave_et470,y_et470,label='EGFP Excitation Filter',linewidth=2,color='blue')
ax2.plot(wave_et525,y_et525,label='EGFP Emission Filter',linewidth=2,color='green')
ax2.plot(wave_t495,y_t495,label='EGFP Beam Splitter',linewidth=4,color='red',linestyle='--')
plt.ylabel('Percentage of Transmission (%T)')
plt.legend()
plt.xlim([380,600])
```

### Conclusion
If we stain nucleus with ECFP while staining endoplasmic reticulum with EGFP, ECFP will interfere seriously with EGFP, which means that when we want to focus on endoplasmic reticulum, nucleus will also appear in the scope. However, we can get a clear vision on nucleus only because EGFP optic path does not interfere with ECFP. Thus we can identify nucleus first, then subtract the nucleus from the organelles to separate endoplasmic reticulum. Nevertheless, the better choice is to select a new type of dye which should be distinct from EGFP significantly.
